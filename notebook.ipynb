```python
import pandas as pd
import seaborn as sns
```

```python
df_white = pd.read_csv("winequality-white.csv")
df_white.head()
```

```python
df_red = pd.read_csv("winequality-red.csv")
df_red.head()
```

```python
sns.set(rc={'figure.figsize':(15,12)})
sns.heatmap(df_white.corr(),annot=True, cmap='RdBu_r')
```

```python
sns.set(rc={'figure.figsize':(15,12)})
sns.heatmap(df_red.corr(),annot=True, cmap='RdBu_r')
```
